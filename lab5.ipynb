```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.axes3d import Axes3D
```

```python
X = np.array([[ 1,  1],
              [ 1,  1],
              [ 1,  2],
              [ 1,  5],
              [ 1,  3],
              [ 1,  0],
              [ 1,  5],
              [ 1, 10],
              [ 1,  1],
              [ 1,  2]])
```

```python
y = [45, 55, 50, 55, 60, 35, 75, 80, 50, 60]
```

```python
def calc_mae(y, y_pred):
    err = np.mean(np.abs(y - y_pred))
    return err

def calc_mse(y, y_pred):
    err = np.mean((y - y_pred)**2)
    return err
```

```python
n = X.shape[0]

eta = 0.06
stop_w = 0.001  # Пороговое значение для остановки
W = np.array([1, 0.5])

print(f'Number of objects = {n} \
       \nLearning rate = {eta} \
       \nInitial weights = {W} \n')

y_pred_list = []
prev_W = W.copy()
iter_count = 0

while True:
    y_pred = np.dot(X, W) #используется матричное умножение X @ W
    err = calc_mse(y, y_pred)

    for k in range(W.shape[0]):
        W[k] -= eta * (1/n * 2 * X[:, k] @ (y_pred - y))

    iter_count += 1

    if iter_count % 10 == 0:
        eta /= 1.1
        print(f'Iteration #{iter_count}: W_new = {W}, MSE = {round(err, 2)}')
        y_pred_list.append(y_pred)

    if np.linalg.norm(W - prev_W) < stop_w:
        print('End')
        break

    prev_W = W.copy()
```
